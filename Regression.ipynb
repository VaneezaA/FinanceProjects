```python
import numpy as np
import pandas as pd
from scipy import stats
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
# Univariate Regression

X = data['House Size (sq.ft.)']
Y = data['House Price']
```

```python
# Let's run the REGRESSION
```

```python
# We must assign a constant

X1 = sm.add_constant(X)
reg = sm.OLS(Y,X1).fit() #OLS = Ordinary Least Squares Regression

```

```python
reg.summary()
# Our R^2 is 0.678
# That means that our explanatory/ independant variable can explain aboout 68% of the dependant variable
    # We confirm: Out model has good explanatory power
# beta is the slope
```
