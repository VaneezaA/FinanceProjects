```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
```

```python
data_source = 'yahoo'
start = '1995-1-1'
PG = wb.DataReader('PG', data_source , start)
```

```python
PG.head()
```

```python
PG.tail()
```

```python
PG['simple return'] = (PG['Adj Close']/ PG['Adj Close'].shift(1))-1
print (PG['simple return'])
```

```python
PG['simple return'].plot(figsize=(8,5))
plt.show
```

```python
avg_returns_d = PG['simple return'].mean()
avg_returns_d
```

```python
# number of trading days is NOT 365 (holidays and weekends) --> We use 250
avg_returns_a = PG['simple return'].mean()*250
avg_returns_a
```

```python
print (str(round(avg_returns_a,5)*100)+ '%')
```

```python
# Let's compute the LOG returns
# The Formula is:
# Ln (Pt / Pt-1)
```

```python
PG.head()
```

```python
# We call the log function from the Numpy module
PG['log return'] = np.log(PG['Adj Close']/ PG['Adj Close'].shift(1))
print (PG['log return'])
```

```python
PG['log return'].plot(figsize=(8,5))
plt.show
```

```python
log_returns_d = PG['log return'].mean()
log_returns_d
```

```python
# number of trading days is NOT 365 (holidays and weekends) --> We use 250
log_returns_a = PG['log return'].mean()*250
log_returns_a
```

```python
print (str(round(log_returns_a,5)*100)+ '%')
```

```python
# Portfolio Return
```

```python
tickers = ['PG', 'MSFT', 'F', 'GE']
mydata = pd.DataFrame()
for t in tickers:
    mydata[t] = wb.DataReader(t, data_source= 'yahoo', start = '1995-1-1')['Adj Close']
```

```python
# Normalization to 100 = Pt/P0 * 100
(mydata/mydata.iloc[0]*100).plot(figsize = (15,16));
plt.show
```

```python
mydata.plot(figsize = (15,16))
plt.show()
```

```python
# Let's calculate simple returns of the portfolio
```

```python
returns = (mydata / mydata.shift(1))-1
returns.head()
```

```python
weights = np.array([0.25, 0.25, 0.25, 0.25])
```

```python
np.dot(returns,weights)
```

```python
annual_returns = returns.mean() * 250
annual_returns
```

```python
np.dot(annual_returns,weights)
```

```python
pfolio_1 = (str(round(np.dot(annual_returns, weights),5)*100)) + '%'
print (pfolio_1)
```

```python
weights_2 = np.array([0.4, 0.4, 0.15, 0.15])
```

```python
pfolio_2 = (str(round(np.dot(annual_returns, weights_2),5)*100)) + '%'
print (pfolio_1)
print (pfolio_2)
```
